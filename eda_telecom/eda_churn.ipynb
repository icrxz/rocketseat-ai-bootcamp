# Análise bivariada

```python
# Hipótese: Clientes com Contrato do tipo mensal estão mais propensos a abandonar o serviço
# Construir tabela de contingência entre Tipo de Contrato e Churn - Quantidade
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total')
```

```python
# Construir tabela de contingência entre Tipo de Contrato e Churn - Percentual
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total', normalize='index') * 100
```

Considerando o público total, os clientes com contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham o contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal com o abandono do serviço

### Avaliar a correlação entre 2 variaveis categóricas (qualitativas nominais, não possuem ordem)

Executar um teste de hipótese chamado Chi-Square ou Qui-Quadrado de Pearson. (analisa se as variáveis são dependentes ou não)

Num teste de hipótese, duas hipóteses são formuladas:
- H0 (Hipótese nula): as duas variáveis são independentes, não possui correlação
- H1 (Hipótese complementar): as duas variáveis não são independentes

O teste serve para confirmar ou recusar a hipotese nula (H0).

Quando a probabilidade de observamos H0 é inferior ao valor de 0.05 (p-value, valor padrão para análise estátistica, porém pode ser ajustado caso a caso se necessário), recusamos a hipotese nula e seguimos com a complementar


```python
# Gerar tabela de contingência, um DF de crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Values
from scipy.stats import chi2_contingency #É necessário instalar a lib scipy
```

```python
# Cria ção de Tupla (Tupla = Lista imutável)
chi_scores_churn_contact = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contact
```

```python
# Para facilitar, criamos 2 séries com os scores e p-values
scores_churn_contract = pd.Series(chi_scores_churn_contact[0])
pvalues_churn_contract = pd.Series(chi_scores_churn_contact[1])
```

```python
# Apresentar números decimais sem utilizar notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar dataframe com os resultados das series
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Values
df_chi_scores_churn_contract
```

Conforme mostrado pelo teste estatístico (Chi-Square), o resultado do valor de p-value é menor do que 0.05, conforme definido anteriormente podemos rejeitar a hipótese nula (H0), ou seja, as variaveis não são independentes (possui correlação). Pelo Qui2 alto, podemos afirmar que há uma forte correlação.
